```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#Loading the dataset
df=pd.read_csv('parkinsons.csv')
```

```python
df.info()
```

```python
#Changing categorical data into num
le = LabelEncoder()
label = le.fit_transform(df['name'])
label
```

```python
df.drop("name", axis=1, inplace=True)
df["name"] = label
df
```

```python
#Checking null values
df.isnull().sum()
```

```python
#Making Output 
x=df.drop('status',axis=1)
y=df['status']
```

```python
x.columns
```

```python
#Splitting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=33)
```

```python
#Fitting and Predicting
gnb = GaussianNB()
gnb.fit(x_train,y_train)
pred=gnb.predict(x_test)
print('Validation accuaracy score  ',accuracy_score(y_test,pred)*100)
```

```python
#Cross validation
cv_model=GaussianNB()
scores = cross_val_score(cv_model, x, y, cv =10,scoring='accuracy' )
Accuracy_cross_val=(scores.mean())*100
print("Accuracy with 10 k fold validation: ",Accuracy_cross_val)
```
